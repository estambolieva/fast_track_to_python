```python
#GET DATA PROPERLY!!!
from urllib2 import Request, urlopen
import pandas as pd
import sklearn
import numpy as np
import seaborn as sns
from scipy import stats, integrate
import matplotlib.pyplot as plt


# for having the graphs show up in the notebook, instead of as a pop up: 
% matplotlib inline

# Quick hack to avoid filling the screen with boring warnings 
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
request=Request('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data')
```

```python
response = urlopen(request)
```

```python
html=response.read()
```

```python
#df=pd.DataFrame(html,columns = ['age','workClass','fnlwdt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_week','native_country'])
```

```python
lines=html.split('\n')
len(lines)
```

```python
table=[]
for line in lines:
    table.append(line.split(', '))
```

```python
df=pd.DataFrame(table, dtype=str ,columns = ['age','workClass','fnlwdt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_week','native_country','income'])
```

```python
df.head(4)
```

```python
df=df.drop([32561, 32562])
df.shape
```

```python
df[['age','education_num','capital_gain','capital_loss','hours_week']] = df[['age','education_num','capital_gain','capital_loss','hours_week']].apply(pd.to_numeric)
```

```python
df['age']=df['age'].fillna(40)
```

```python
df['income'] = df['income'].map({'<=50K': 0, '>50K': 1})
```

```python
criterionF = df['sex'].map(lambda x: x.startswith('F'))
criterionM = df['sex'].map(lambda x: x.startswith('M'))
criterionWhite = df['race'].map(lambda x: x.startswith('White'))
criterionBlack = df['race'].map(lambda x: x.startswith('Black'))
criterionAsian = df['race'].map(lambda x: x.startswith('Asian'))
criterionEskimo = df['race'].map(lambda x: x.startswith('Amer'))
criterionOther = df['race'].map(lambda x: x.startswith('Other'))
```

```python
df.race.value_counts()
```

```python
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(ncols=5, sharey=True, figsize=(15, 7))
a=sns.distplot(df[criterionWhite]['income'], kde=False,ax=ax1,bins=3,norm_hist=True)
a.set(xlabel='<50K                        >=50K')
a.set(ylabel='White')

b=sns.distplot(df[criterionBlack]['income'], kde=False,ax=ax2,bins=3,norm_hist=True)
b.set(xlabel='<50K                        >=50K')
b.set(ylabel='Black')

c=sns.distplot(df[criterionAsian]['income'], kde=False,ax=ax3,bins=3,norm_hist=True)
c.set(xlabel='<50K                        >=50K')
c.set(ylabel='Asian-Pac-Islander')

d=sns.distplot(df[criterionEskimo]['income'], kde=False,ax=ax4,bins=3,norm_hist=True)
d.set(xlabel='<50K                        >=50K')
d.set(ylabel='Amer-Indian-Eskimo')

e=sns.distplot(df[criterionOther]['income'], kde=False,ax=ax5,bins=3,norm_hist=True)
e.set(xlabel='<50K                        >=50K')
e.set(ylabel='Other')
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
g=sns.distplot(df[criterionM]['income'], kde=False,ax=ax1,bins=2,norm_hist=True)
g.set(xlabel='<50K             >=50K')
g.set(ylabel='Males')

h=sns.distplot(df[criterionF]['income'], kde=False,ax=ax2,bins=2,norm_hist=True)
h.set(xlabel='<50K             >=50K')
h.set(ylabel='Females')
```
