PDB defines an interactive source code debugger for Python programs.

```python
import pdb
```

```python
def greet(hour):
    if hour < 12:
        print 'Good morning!'
    elif hour >= 12 and hour < 20:
        print 'Good afternoon!'
    else:
        pdb.set_trace()
```

```python
greet(11)
```

```python
greet(14)
```

```python
greet(50)
```
