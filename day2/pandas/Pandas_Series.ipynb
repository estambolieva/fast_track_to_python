### Series - an indexed 1D array

```python
s = pd.Series([1,3,5,np.nan,6,8])
s
```

```python
type(s.index)
```

```python
# from a dict
d = {'a' : 0., 'b' : 1., 'c' : 2.}
pd.Series(d)
```

```python
pd.Series(d, index=['b', 'c', 'd', 'a'])
```

```python
pd.Series(5., index=['a', 'b', 'c', 'd', 'e'])
```
